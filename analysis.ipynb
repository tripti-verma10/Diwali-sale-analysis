**EXPORATARY DATA ANALYSIS**

GENDER

```python
print(df.columns)
```

```python
ax= sns.countplot(x="Gender" , data=df,palette="coolwarm" ,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title("Gender Distribution", fontsize=14, fontweight='bold')
    plt.xlabel("Gender", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)


```

```python
sales = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender' , y='Amount', data=sales ,hue='Gender')
plt.title("Total Sales Amount by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Total Amount", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

**AGE**

```python
print(df.columns)
```

```python
ax= sns.countplot(x="Age Group" , data=df,palette="coolwarm" ,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Age Group Distribution by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
sale_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group' , y='Amount', data=sale_age)
plt.title("Total Sales Amount by Age Group", fontsize=14, fontweight='bold')
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Total Amount Spent", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

**state**

```python
print(df.columns)
```

```python
sale_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State' , y='Orders', data=sale_state)
plt.title("Total Orders by State", fontsize=14, fontweight='bold')
plt.xlabel("State", fontsize=12)
plt.ylabel("Total Orders", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

```

```python
sale_amount= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State' , y='Amount', data=sale_amount)
plt.title("Total Amount Spent by State", fontsize=14, fontweight='bold')
plt.xlabel("State", fontsize=12)
plt.ylabel("Total Amount", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
